```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sb
sb.set()
```

```python
raw_data= pd.read_csv('real_estate_price_size_year_view.csv')
raw_data
```

```python
data=raw_data.copy()
data['view']=data['view'].map({'No sea view':0,'Sea view': 1})
```

```python
y=data['price']
x1=data[['size','year','view']]
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['size'],y)
plt.xlabel('Dependencies', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```

```python
new_data=pd.DataFrame({'const':1,'size':[1000,500], 'year':[2008,2019], 'view':[0,1]})
new_data
```

```python
new_data=new_data.rename(index={0:'house1', 1:'house2'})
new_data
```

```python
predictions=result.predict(new_data)
predictions
```

```python
predictiondf=pd.DataFrame({'price prediction':predictions})
joined=new_data.join(predictiondf)
joined
```
