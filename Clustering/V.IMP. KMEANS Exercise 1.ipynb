```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
seaborn.set()
from sklearn.cluster import KMeans
```

```python
raw_data=pd.read_csv('iris-dataset.csv')
raw_data
```

```python
plt.scatter(raw_data['sepal_length'],raw_data['sepal_width'])
plt.show()
```

```python
#KMeans without standardizing
x=raw_data.copy()
kmeans=KMeans(2)
kmeans.fit(x)
```

```python
#adding column to data
clusters=raw_data.copy()
clusters['cluster']=kmeans.fit_predict(x)
clusters
```

```python
plt.scatter(clusters['sepal_length'],clusters['sepal_width'], c=clusters['cluster'],cmap='rainbow')
plt.show()
```

```python
#Standardizing variables
from sklearn import preprocessing
scaled=preprocessing.scale(raw_data)
```

```python
kmeans_scaled=KMeans(2)
kmeans_scaled.fit(scaled)
```

```python
#Adding Clusters column
scaled=raw_data.copy()
scaled['clusters']=kmeans_scaled.fit_predict(scaled)
scaled
```

```python
plt.scatter(scaled['sepal_length'], scaled['sepal_width'], c= scaled ['clusters'], cmap = 'rainbow')
plt.show()
```
